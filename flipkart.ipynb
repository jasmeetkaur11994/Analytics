```python
#by Jasmeet Kaur
#Answer 1
from bs4 import BeautifulSoup as bs
import requests
import csv
import pandas as pd
```

```python
#searching mobile product
link='https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
page = requests.get(link)
```

```python
#creating beautiful soup object
soup = bs(page.content, 'html.parser')
#the following will give us the visual representation of data
print(soup.prettify())
```

```python
product=[]
prices=[]
ratings=[]
url=[]
```

```python
for a in soup.findAll('a', href=True,attrs={'class':'_1fQZEK'}):
    name=a.find('div',attrs={'class':'_4rR01T'})
    price=a.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
    rating=a.find('div',attrs={'class':'_3LWZlK'})
    urls=a.attrs.get('href') 
    product.append(name.text)
    prices.append(price.text)
    ratings.append(rating.text)
    url.append(urls)
```

```python
#checking size of all objects before creating dataframe
print(len(product))
print(len(ratings))
print(len(prices))
print(len(url))
```

```python
df=pd.DataFrame({'Product Name':product,'Price':prices,'Rating':ratings,'URL':url})
df.head(10)
```

```python
df.to_csv('Flipkart_mobile')
```
